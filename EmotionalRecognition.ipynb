```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
a1 = cv2.imread("maara.jpg")
plt.imshow(a1[:,:,::-1])
plt.show()
```

```python
res = DeepFace.analyze(a1, actions = ['emotion'])
```

```python
print(res)
```

```python
print(res['dominant_emotion'])
```

```python
def dominant(z):

  if z == 'angry':
    a1 = angry
    return(a1)

  elif z == 'disgust':
    a1 = disgust
    return(a1)
  elif z == 'fear':
    a1 = fear
    return(a1)

  elif z == 'happy':
    a1 = happy
    print("Life is Short Stay Happy Stay the same bruh")
    return(a1)
    
  elif z == 'sad':
    a1 = sad
    print("Get motivated with this Song:")
    return(a1)
    
  elif z == 'surprise':
    a1 = surprise
    return(a1)
  else:
    a1 = neutral
    return(a1)
```

```python
dominant(res['dominant_emotion'])
```
